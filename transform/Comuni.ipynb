```python
import pandas as pd
import geopandas as gpd
```

```python
path_regioni = f'../dataset/dati_spaziali/regioni/regioni.shp'
path_comuni = f'../dataset/dati_spaziali/comuni/comuni.shp'
```

```python
comuni = gpd.read_file(path_comuni)
```

```python
comuni.head()
```

```python
regioni = gpd.read_file(path_regioni)
```

Estrazione dei comuni pugliesi

```python
codice_puglia = regioni[regioni['DEN_REG'] == 'Puglia']['COD_REG'].values[0]
comuni_puglia = comuni[comuni['COD_REG'] == codice_puglia]
comuni_puglia.reset_index(inplace=True, drop=True)
```

```python
comuni_puglia.head()
```

```python
comuni_puglia = comuni_puglia[['COD_PROV', 'PRO_COM', 'COMUNE', 'geometry']]
comuni_puglia.rename(columns={'COMUNE': 'nome',
                              'COD_PROV': 'id_prov',
                              'PRO_COM': 'id_comune'}, 
                       inplace=True)
```

```python
comuni_puglia.head()
```

```python
comuni_puglia.to_file('../dataset/dati_spaziali/comuni/comuni_finale.shp', index=None)
```
