```python
import pandas as pd
import geopandas as gpd
```

```python
path_decessi = '../dataset/decessi/decessi_malattie_2019_2020.csv'
path_province = f'../dataset/dati_spaziali/province/province.shp'
```

```python
province = gpd.read_file(path_province)
```

```python
province.head()
```

```python
decessi = pd.read_csv(path_decessi)
```

```python
decessi.head()
```

```python
columns = ['Territorio', 'Sesso', 'Causa iniziale di morte - European Short List', 'TIME', 'Value']
decessi = decessi[columns]
```

```python
decessi
```

Creazione del dataframe con colonne 'nome_malattia', 'numero_decessi_maschi', 'numero_decessi_femmine', 'anno', 'id_provincia'

```python
decessi_finale = pd.DataFrame(columns = ['nome_malattia', 'numero_decessi_maschi', 'numero_decessi_femmine', 'anno', 'id_provincia'])
```

```python
for p in set(decessi['Territorio']):
    decessi_territorio = decessi[decessi['Territorio'] == p]
    for c in set(decessi_territorio['Causa iniziale di morte - European Short List']):
        decessi_territorio_malattia = decessi_territorio[decessi_territorio['Causa iniziale di morte - European Short List'] == c]
        anni = set(decessi['TIME'])
        for a in anni:
            decessi_territorio_malattia_maschi = decessi_territorio_malattia[decessi_territorio_malattia['Sesso'] == 'maschi']
            decessi_territorio_malattia_maschi_anno = decessi_territorio_malattia_maschi[decessi_territorio_malattia_maschi['TIME'] == a]['Value'].values[0]
            
            decessi_territorio_malattia_femmine = decessi_territorio_malattia[decessi_territorio_malattia['Sesso'] == 'femmine']
            decessi_territorio_malattia_femmine_anno = decessi_territorio_malattia_femmine[decessi_territorio_malattia_femmine['TIME'] == a]['Value'].values[0]
            id_provincia = province[province['DEN_UTS'] == p]['COD_PROV'].values[0]
            
            row = [c, decessi_territorio_malattia_maschi_anno, decessi_territorio_malattia_femmine_anno, a, id_provincia]
            decessi_finale.loc[len(decessi_finale)] = row
```

```python
decessi_finale.to_csv('..\dataset\decessi\decessi_finale.csv', index=False)
```
