```python
import pandas as pd
import geopandas as gpd
```

```python
path_regioni = f'../dataset/dati_spaziali/regioni/regioni.shp'
path_province = f'../dataset/dati_spaziali/province/province.shp'
```

```python
regioni = gpd.read_file(path_regioni)
province = gpd.read_file(path_province)
```

```python
regioni.head()
```

```python
province.head()
```

Vengono considerate solo le province della Puglia

```python
codice_puglia = regioni[regioni['DEN_REG'] == 'Puglia']['COD_REG'].values[0]
```

```python
province_puglia = province[province['COD_REG'] == codice_puglia]
province_puglia.reset_index(inplace=True, drop=True)
```

```python
province_puglia = province_puglia[['DEN_UTS', 'COD_PROV', 'geometry']]
province_puglia.rename(columns={'DEN_UTS': 'nome', 
                                'COD_PROV': 'id_prov',
                                'geometry': 'geom'}, 
                       inplace=True)
```

```python
province_puglia.head()
```

```python
province_puglia.to_file('../dataset/dati_spaziali/province/province_finale.shp', index=None)
```
